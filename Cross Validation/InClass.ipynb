## LOOCV

```python
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
accuracy = 0.0
for i in range(X.shape[0]):
    x_train = np.delete(X, i, axis = 0)
    y_train = np.delete(Y, i)
    
    x_test = X[i,:]
    y_test = Y[i]
    
    classifier = LogisticRegression()
    classifier.fit(x_train, y_train)
    
    y_pred = classifier.predict(x_test.reshape(1, -1))
    cm = confusion_matrix(y_test, y_pred)

    count = 0
    index = 0
    for pred in y_pred:
        if pred == y_test[index]:
            count = count + 1
        else:
            pass
        index = index + 1
    accuracy = accuracy + count/len(y_pred)

totalAccuracy = accuracy/X.shape[0]
```

## K-Fold

```python
def kFoldMethod(X, Y, kFold):
    foldTimes = len(X)/kFold
    accuracy = 0.0
    for n in (np.arange(foldTimes)):
        rowIndicesToRemove = np.arange(n*kFold,(n+1)*kFold)
        rowIndicesToRemove = [ int(x) for x in rowIndicesToRemove ]

        x_train = np.delete(X,rowIndicesToRemove,axis = 0)
        y_train = np.delete(Y,rowIndicesToRemove,axis = 0)
        
        x_test = X[rowIndicesToRemove[0]:rowIndicesToRemove[kFold-1]+ 1,:]
        y_test = Y[rowIndicesToRemove[0]:rowIndicesToRemove[kFold-1]+ 1]
    
        classifier = LogisticRegression(solver = 'lbfgs', multi_class = 'auto')
        classifier.fit(x_train, y_train)
    
        y_pred = classifier.predict(x_test)
        numberOfTrues = 0
        for index, pred in enumerate(y_pred):
            if pred == y_test[index]:
                numberOfTrues = numberOfTrues + 1
            else:
                pass
        accuracy = accuracy + numberOfTrues/len(y_pred) # Divide number of trues to number of elements in y_pred to get accuracy for each prediction
    totalAccuracy = accuracy/foldTimes
    return totalAccuracy
```

```python
kFoldMethod(X,Y,1)
```
